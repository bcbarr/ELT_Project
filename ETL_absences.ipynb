```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#data collected from https://www.cde.ca.gov/ds/sd/sd/filesabd.asp

text_file = "chronic_absentee_1819.txt"
absence_data_df = pd.read_table(text_file, encoding="latin-1")
absence_data_df.head(30)
```

```python
#Clean the Data
asian = absence_data_df.loc[(absence_data_df["ReportingCategory"]=="RA") & (absence_data_df["CharterYN"]=="All")]

asian.head()
clean_asian = asian[["CountyName", "CumulativeEnrollment", "ChronicAbsenteeismEligibleCumulativeEnrollment","ChronicAbsenteeismRate"]]
clean_asian
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/"
engine = create_engine(f'postgresql://{rds_connection_string}')
```
